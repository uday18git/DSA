```python
#nqueen

def isSafe(arr,x,y,n):
    for i in range(x): # in the column column 
        if arr[i][y]==1:
            return False
    col=y
    row=x
    while row>=0 and col>=0: #left diagonal
        if arr[row][col]==1:
            return False
        row-=1
        col-=1
    col=y
    row=x
    while row>=0 and col<n: #right diagonal
        if arr[row][col]==1:
            return False
        row-=1
        col+=1
    return True
def nqueen(arr,x,n):
    if(x>=n):
        return True
    for col in range(n):
        if isSafe(arr,x,col,n):
            arr[x][col]=1
            if nqueen(arr,x+1,n):
                return True
            arr[x][col]=0
    return False


if __name__ == "__main__":
    n = int(input("Enter size"))
    arr= [[0 for _ in range(n)]for _ in range(n)] 
    if nqueen(arr,0,n):
        for i in range(n):
            for j in range(n):
                if arr[i][j]==1:
                    print('  Q  ',end='')
                else:
                    print('  .  ',end='')
            print('\n')
    else:
        print("No soln found")
            
```

```python
#B9 Travelling Salesman

def getpermutations(arr,l,r):
    if l==r:
        return [arr[:]]
    else:
        permutations=[]
        for i in range(l,r+1):
            arr[i],arr[l]=arr[l],arr[i]
            permutations.extend(getpermutations(arr,l+1,r))
            arr[l],arr[i]=arr[i],arr[l]  #backtracking
        return permutations


def tsp(graph,start,cities):
    vertex=[i for i in range(cities) if i!=start]
    p=getpermutations(vertex,0,len(vertex)-1)                # l is lower index , r is upper index of list to be permutated
    M=[]
    
    for i in p:
        d=0
        k=start
        flag=0
        
        for j in i:
            if graph[k][j]==0:
                break
            else:
                d+=graph[k][j]
                k=j
                flag+=1
        if flag==4:
            if graph[k][start]:
                d+=graph[k][start]
                M.append(d)
    M.sort()
    return(M[0])

graph=[[0,2,0,12,5],
      [2,0,4,8,0],
      [0,4,0,3,3],
      [12,8,3,0,10],
      [5,0,3,10,0]]
cities=5
start=int(input("Enter Starting city : A:0 B:1 C:2 D:3 E:4 :"))
a=tsp(graph,start,cities)
print("Minimum Distance is : ",a)
```

```python
from sys import maxsize
v=4
def travelling_salesman_function(graph,s): #to calculate the path
    vertex = []
    for i in range(v):
        if i!=s:
            vertex.append(i)
    min_path = maxsize
    while True:
        current_cost=0 
        k=s
        for i in range(len(vertex)):
            current_cost += graph[k][vertex[i]]
            k=vertex[i]
        current_cost+=graph[k][s] # after ending for loop add the cost of going frm the current index to the sourcee because we have to make a loop
        min_path =min(min_path,current_cost)

        if next_perm(vertex)==False:
            break
    return min_path

def next_perm(l):
    n=len(l)
    i=n-2
    while i>=0 and l[i]>l[i+1]:
        i-=1
    if i==-1:
        return False
    
    j=i+1
    while j<n and l[j]>l[i]:
        j+=1
    j-=1
    l[i],l[j]=l[j],l[i]
    left=i+1
    right=n-1
    
    while left<right:
        l[left],l[right] = l[right],l[left]
        left+=1
        right-=1
    return True
graph = [[0,10,15,20],[10,0,35,25],[15,35,0,30],[20,25,30,0]]
s=0
print(travelling_salesman_function(graph,s))
```
